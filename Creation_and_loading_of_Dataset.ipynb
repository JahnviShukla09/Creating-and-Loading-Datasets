```python
import pandas as pd
data = {
    "Roll No.": [ 101,102,103,104,105],
    "Gender": ["Female", "Male", "Female","Male","Female"],
    "Department": ["Computer", "IT","EnTC","Mechanical","Civil"],
    "CGPA":[8.2,7.5,9.1,6.8,8.7]
}
df = pd.DataFrame(data)
df
```

```python
df.to_csv('students.csv', index=False) #False so that the data does not change
```

```python
# Perform mean, median, mode operation
print('Mean of CGPA:', df['CGPA'].mean())
print('Median of CGPA:', df['CGPA'].median())
print('Mode of CGPA:', df['CGPA'].mode()[0])
```

```python
# Insert a new column Grade
df["Grade"] = ["First Class", "Distinction", "First Class", "Second Class", "Second Class"]
df
```

```python
# Delete Gender column
df = df.drop(columns=['Gender'])
display(df)
```

```python
# CGPA of Topper
print('Topper CGPA:', df['CGPA'].max())
```
